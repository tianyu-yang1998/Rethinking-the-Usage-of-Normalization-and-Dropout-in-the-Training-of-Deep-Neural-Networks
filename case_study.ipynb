```python
def load_model(model, state_dict_path, print_layers=False):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    state_dict = torch.load(state_dict_path, map_location=device)
    model.load_state_dict(state_dict)
    if print_layers:
        for key in state_dict.keys():
            print(key)
    return model
```

```python
model_resnet110 = load_model(
    resnet.resnet110(num_classes=10),
    "./res/cifar10_resnet110_best.pth",
    True, 
)

model_resnet110_ic = load_model(
    resnet.resnet110_ic(num_classes=10),
    "./res/cifar10_resnet110_ic_best.pth",
    True, 
)
```
