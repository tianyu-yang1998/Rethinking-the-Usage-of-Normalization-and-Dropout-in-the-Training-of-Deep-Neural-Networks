```python
plt.figure(figsize=[16,8], dpi=200)
models = ['vgg16', 'googlenet', 'resnet110']
datasets = ['cifar10', 'cifar100']

for i, model in enumerate(models):
    for j, dataset in enumerate(datasets):
        baseline_file = f"./res/{dataset}_{model}.npy"
        ic_file = f"./res/{dataset}_{model}_ic.npy"
        if os.path.exists(baseline_file):
            res = np.load(baseline_file)
        else:
            res = []
        if os.path.exists(ic_file):
            res_ic = np.load(ic_file)
        else:
            res_ic = []
    
        plt.subplot(2,3,j*3+i+1)
        plt.plot(res)
        plt.plot(res_ic)
        
        if j == 0:
            plt.ylim([75,100])
            plt.xticks([])
        else:
            plt.ylim([50,75])
            plt.xlabel('Epoch')
        
        plt.title(model)
        plt.ylabel(dataset)
        plt.legend(['Basline', 'IC'])
        plt.savefig('./figures/final_res.png', bbox_inches='tight')
            
        
```
